## Vistas
Permiten almacenar una consulta para usarla despues.

### Ventajas
* Oculta informaciÃ³n
* Simplificar permisos
* ReutilizaciÃ³n

### Desventajas
* No se puede incluir **compute, into**
* No se pueden basar en tablas temporales
* No se pueden asociar reglas ni defaults

Nota: No se puede insertar a una vista si esta depende de mÃ¡s de una tabla.

Nota: Se respectan los constraints de las tablas de las que depende.

```sql
if object_id('empleados') is not null
  drop table empleados;
 if object_id('secciones') is not null
  drop table secciones;

 create table secciones(
  codigo tinyint identity,
  nombre varchar(20),
  sueldo decimal(5,2)
   constraint CK_secciones_sueldo check (sueldo>=0),
  constraint PK_secciones primary key (codigo)
 );

 create table empleados(
  legajo int identity,
  documento char(8)
   constraint CK_empleados_documento check (documento like '[0-9][0-9][0-9][0-9][0-9][0-9][0-9][0-9]'),
  sexo char(1)
   constraint CK_empleados_sexo check (sexo in ('f','m')),
  apellido varchar(20),
  nombre varchar(20),
  domicilio varchar(30),
  seccion tinyint not null,
  cantidadhijos tinyint
   constraint CK_empleados_hijos check (cantidadhijos>=0),
  estadocivil char(10)
   constraint CK_empleados_estadocivil check (estadocivil in ('casado','divorciado','soltero','viudo')),
  fechaingreso datetime,
   constraint PK_empleados primary key (legajo),
  constraint FK_empleados_seccion
   foreign key (seccion)
   references secciones(codigo)
   on update cascade,
  constraint UQ_empleados_documento
   unique(documento)
);

 insert into secciones values('Administracion',300);
 insert into secciones values('ContadurÃ­a',400);
 insert into secciones values('Sistemas',500);

 insert into empleados values('22222222','f','Lopez','Ana','Colon 123',1,2,'casado','1990-10-10');
 insert into empleados values('23333333','m','Lopez','Luis','Sucre 235',1,0,'soltero','1990-02-10');
 insert into empleados values('24444444','m','Garcia','Marcos','Sarmiento 1234',2,3,'divorciado','1998-07-12');
 insert into empleados values('25555555','m','Gomez','Pablo','Bulnes 321',3,2,'casado','1998-10-09');
 insert into empleados values('26666666','f','Perez','Laura','Peru 1254',3,3,'casado','2000-05-09');

 if object_id('vista_empleados') is not null
  drop view vista_empleados;
 

/*Creamos la vista "vista_empleados",
que es resultado de una combinaciÃ³n en la cual se muestran 4 campos 
(nombre completo, sexo, seccion, cantidad de hijos*/
GO
CREATE VIEW vista_empleados
AS 
(
  SELECT E.nombre + '' + E.apellido AS nombre_completo, sexo, S.nombre as seccion, E.cantidadhijos FROM empleados E INNER JOIN secciones S ON  E.seccion = S.codigo
);
GO

/* Ver la tabla */
SELECT * FROM vista_empleados;

/* Ver informaciÃ³n de tabla */
EXEC sp_help vista_empleados;

/* Ver definiciÃ³n de tabla*/
EXEC sp_helptext vista_empleados;

/* Ver dependencias de vista */
EXEC sp_depends vista_empleados;

/* Eliminar la vista y volverla a crear usando la opciÃ³n with encryption */
DROP VIEW vista_empleados;
GO
CREATE VIEW vista_empleados
WITH ENCRYPTION
AS
(
  SELECT
    E.nombre + ' ' + E.apellido AS nombre_completo,
    sexo,
    S.nombre AS seccion,
    E.cantidadhijos
  FROM empleados E INNER JOIN secciones S ON E.seccion = S.codigo
)
```

```sql
-- Creamos el procedimiento para que reciba el nombre de un autor y 
-- muestre todos los libros del autor solicitado:
CREATE PROCEDURE get_libros
@autor VARCHAR(100)
AS
-- BEGIN y END son opcionales
BEGIN
    SELECT * FROM libros WHERE autor LIKE '%'+ @autor + '%';
END

GO

-- Ejecutamos el procedimiento:
EXEC get_libros 'Borges';

-- Empleamos la otra sintaxis (por nombre):
EXEC get_libros @autor='Borges';

-- Eliminar procedimiento
DROP PROC get_libros;
GO
```

```sql
-- Creamos un procedimiento "pa_libros_autor_editorial" que recibe 2 parÃ¡metros,
-- el nombre de un autor y el de una editorial:
CREATE PROCEDURE pa_libros_autor_editorial
    @autor VARCHAR(100) = 'Borges',
    @editorial VARCHAR(100) = null
AS
BEGIN
    IF @editorial is null
    BEGIN
        SELECT * FROM libros WHERE autor LIKE '%'+ @autor + '%';
    END
    ELSE
    BEGIN
        SELECT * FROM libros WHERE autor LIKE '%'+ @autor + '%' AND editorial LIKE '%'+ @editorial + '%';
    END
END
GO

-- Ejecutamos el procedimiento enviando los parÃ¡metros por posiciÃ³n:
EXEC pa_libros_autor_editorial 'Planeta','Bach';

-- Ejecutamos el procedimiento enviando otros valores y lo hacemos por nombre 
--(Si ejecutamos el procedimiento omitiendo los parÃ¡metros, aparecerÃ¡ un mensaje de error.):
EXEC pa_libros_autor_editorial @editorial='Planeta', @autor='Bach';
GO
```

### Crear procedimiento donde retornamos un valor

**Nota: Usar output o no se nos va a pasar el valor**

```sql
-- Retornar valor del procedimiento
CREATE PROCEDURE suma_simple
@my_num1 INT = 0,
@my_num2 INT = 0,
@answer INT output
AS
BEGIN
    SET @answer = @my_num1 + @my_num2
END
GO

DECLARE @my_var int;
EXEC suma_simple  2, 3, @my_var output;
SELECT @my_var;
GO
```

```sql
-- Otro procedimiento. Notese que en este estamos asignando una variable
CREATE PROCEDURE suma_autor
@num1 int = 0,
@autor VARCHAR(100) = 'Borges',
@answer int output
AS
BEGIN
    DECLARE @num_libros int
    SELECT @num_libros=COUNT(*) FROM libros WHERE autor LIKE '%'+ @autor + '%';
    SET @answer = @num_libros + @num1
END
GO

DECLARE @my_var int;
EXEC suma_autor 4, @answer = @my_var output;
SELECT @my_var;
GO
```

```sql
-- Creamos una funciÃ³n a la cual le enviamos una fecha (de tipo varchar),
-- en el cuerpo de la funciÃ³n se analiza si el dato enviado corresponde a una fecha,
-- si es asÃ­, se almacena en una variable el mes (en espaÃ±ol) y se le concatenan el dÃ­a y el aÃ±o
-- y se retorna esa cadena; en caso que el valor enviado no corresponda a una fecha,
-- la funciÃ³n retorna la cadena 'Fecha invÃ¡lida':

CREATE FUNCTION fecha_esp(@fecha_str VARCHAR(100)) returns VARCHAR(100)
AS
BEGIN
    DECLARE @answer VARCHAR(100)
    SET @answer = 'Fecha invalida'
    IF ISDATE(@fecha_str) = 1
    BEGIN
        DECLARE @day VARCHAR(4)
        DECLARE @month VARCHAR(20)
        DECLARE @year VARCHAR(4)

        DECLARE @fecha DATETIME
        SET @fecha = CAST(@fecha_str AS datetime)

        SET @day = DATEPART(DAY, @fecha)
        SET @year = DATEPART(YEAR, @fecha)
        SET @month = 
            CASE DATEPART(MONTH, @fecha)
                WHEN 1 THEN 'Enero'
                WHEN 2 THEN 'Febrero'
                WHEN 3 THEN 'Marzo'
                WHEN 4 THEN 'Abril'
                WHEN 5 THEN 'Mayo'
                WHEN 6 THEN 'Junio'
                WHEN 7 THEN 'July'
                WHEN 8 THEN 'Agosto'
                WHEN 9 THEN 'Septiembre'
                WHEN 10 THEN 'Octubre'
                WHEN 11 THEN 'Noviembre'
                WHEN 12 THEN 'December'
            END

        SET @answer = @day + ' de ' + @month + ' del ' + @year
    END

    RETURN @answer
END
GO

-- Recuperamos los registros de "empleados", mostrando el nombre y
-- la fecha de ingreso empleando la funciÃ³n creada anteriormente:
SELECT nombre, dbo.fecha_esp(fechaingreso) FROM empleados;
```

## Triggers

Se ejecutan cuando ocurre algo en la base de datos. Ej. inserciÃ³n, update, delete.

Creamos un disparados para que no permita eliminar mÃ¡s de un registro a la vez de la tabla empleados:

```sql
CREATE TRIGGER trigger_limite_borrado_emp
ON empleados
FOR DELETE
AS
BEGIN
    IF(SELECT COUNT(*) FROM deleted) > 1
    BEGIN
        RAISERROR('No se puede borrar mÃ¡s de un registro a la vez', 11, 1)
        ROLLBACK TRANSACTION
    END
END
```

Probar trigger

```sql
DELETE FROM empleados WHERE domicilio LIKE 'Bulnes%';
```

Deshabiltar trigger

```sql
ALTER TABLE empleados DISABLE TRIGGER trigger_limite_borrado_emp;
```

Creamos un disparados para que no permite actualizar el campo "documento" de la tabla empleados:

```sql
CREATE TRIGGER trigger_restrict_doc_update
ON empleados
FOR UPDATE
AS
BEGIN
    IF UPDATE(documento)
    BEGIN
        RAISERROR('No se puede actualizar este campo', 11, 1)
        ROLLBACK TRANSACTION
    END
END
```

Creamos un disaparador que no permita ingresar empleados en la secciÃ³n "Gerencia":

```sql
CREATE TRIGGER trigger_restrict_gerencia_insert
ON empleados
FOR INSERT
AS
BEGIN
    IF (SELECT seccion FROM inserted) = 'Gerencia'
    BEGIN
    RAISERROR('No se puede insertar en la secciÃ³n gerencia', 11, 1)
        ROLLBACK TRANSACTION
    END 
END
```

Probar trigger

```sql
-- Esto deberÃ­a de funcionar
UPDATE empleados SET seccion='Limpieza' WHERE documento='23333333';
```

```sql
-- Esto deberÃ­a fallar
INSERT INTO empleados VALUES ('288888888', 'Juan Juarez', 'Jamaica 123', 'Gerencia');
```

## Transacciones

```sql
CREATE PROC mu_proc
AS 
BEGIN
    BEGIN TRANSACTION T1
        BEGIN TRY
        DECLARE @num_p INT
        INSERT INTO test(pokemon) VALUES ('pikachu');
        SELECT @num_p = COUNT(*) FROM test;

        --- TRANSACCION ANIDADA
        --- Esta transaccion ocurre hasta la que resulte correcta la transaccion 1
        --- Si no se hace rollback si falla T1
        --- No se guarda en la base de datos hasta que T1 haga commit
        BEGIN TRANSACTION T2
            INSERT INTO test(pokemon) VALUES ('pikachu');
            COMMIT TRANSACTION T2

        IF @num_p < 10
        BEGIN
            COMMIT TRANSACTION T1
        END
        ELSE
        BEGIN
            ROLLBACK TRANSACTION T1
        END
    END TRY
    BEGIN CATCH

    END CATCH
END

SELECT * FROM test
```
